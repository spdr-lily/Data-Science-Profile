# configuração do ambiente e criação do dataframe

```python
# importação da biblioteca
import pandas as pd
```

```python
# dados fictícios
data = {
    'Data': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Produto': ['A', 'B', 'A', 'C', 'B'],
    'Quantidade Vendida': [10, 20, 15, 30, 25],
    'Receita': [100, 150, 120, 200, 180],
    'Custo': [80, 120, 90, 180, 150]
}
df = pd.DataFrame(data)
```

```python
# adicionando uma nova coluna chamada Lucro, calculando a direfença entre a receita e o custo
df['Lucro'] = df['Receita'] - df['Custo']
print(df)
```

# Análise Descritiva e Transformação de Dados:

```python
# calcular o total de receita, cs=usto e lucro para todo o período usando as frunções de agregação do Pandas.

# Agragação dos dados
total_receita = df['Receita'].sum()
total_custo = df['Custo'].sum()
total_lucro = df['Lucro'].sum()
print(f'Total de Receita: {total_receita}')
print(f'Total de Custo: {total_custo}')
print(f'Total de Lucro: {total_lucro}')
```

```python
# criar uma coluna que indica a margem de lucro percentual, (margem de lucro (%)), calculada como (Lucro / Receita) * 100.

# margem de lucro percentual
df['Margem de Lucro (%)'] = (df['Lucro'] / df['Receita']) * 100
print(df)
```

```python
# filtrar o DataFrame para exibir apenas os produtos com uma margem de lucro acima de 30%.

produtos_lucrativos = df[df['Margem de Lucro (%)'] > 30]
print(produtos_lucrativos)
```

# Visualização de Dados:

```python
# utilizar bibliotecas matplotlib ou seaborn para criar um gráfico de barras que mostre a quantidade vendida de cada produto.

import matplotlib.pyplot as plt

# gráfico de barras - quantidade vendida por produto
plt.bar(df['Produto'], df['Quantidade Vendida'])
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.title('Quantidade Vendida de Cada Produto')
plt.show()
```

```python
# criar um gráfico de linha que exiba a evolução do lucro ao longo dos dias.

plt.plot(df['Data'], df['Lucro'])
plt.xlabel('Data')
plt.ylabel('Lucro')
plt.title('Evolução do Lucro ao Longo dos Dias')
plt.xticks(rotation=45)
plt.show()
```

```python
# criar um gráfico de pizza que mostra a participação percentual de cada produto na receita total.

plt.pie(df['Receita'], labels=df['Produto'], autopct='%1.1f%%')
plt.title('Participação Percentual de Cada Produto na Receita Total')
plt.show()
```

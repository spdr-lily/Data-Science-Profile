```python
# importação das bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

```python
# criação do conjunto de dados
x = np.linspace(0, 10, 100) # 100 valores entre 0 e 10
y1 = np.sin(x) # valores da função seno para cada valor de x
y2 = np.cos(x) # valores da função cosseno para cada valor de x

# criação de um gráfico de linha para a função seno
plt.plot(x, y1, label='Seno', color='b')
plt.title('Gráfico de Linha: Função Seno') # personalização do gráfico com título
# rótulos para os eixos
plt.xlabel('x')
plt.ylabel('Seno(x)')

plt.show()
```

```python
# Gráfico de Dispersão
plt.scatter(x, y2, label='Cosseno', color='g')
# personalização do gráfico com título e rótulos para os eixos
plt.title('Gráfico de Dispersão: Função Cosseno')
# Personalizando os rótulos
plt.xlabel('x')
plt.ylabel('Cosseno(x)')

plt.show()
```

```python
# gráfico de barras
valores = [2,5,3,7,1] #novo conjunto de dados de valores inteiros
rotulos = ['A', 'B', 'C', 'D', 'E'] # lista de rótulos para barras correspondente aos valores

# criação do gráfico
plt.bar(rotulos, valores, color='lightblue')
plt.title('Gráfico de Barras')
plt.xlabel('Rótulos')
plt.ylabel('Valores')

# salvando o gráfico
plt.savefig('grafico_barras.png', format='png', dpi=300, bbox_inches='tight')
```

```python
# criação de supblots empilhados verticalmente
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Primeiro subplot: gráfico de linha
ax1.plot(x, y1, color='b')
ax1.set_title('Subplot 1: Função Seno')
ax1.set_xlabel('x')
ax1.set_ylabel('Seno(x)')

# Segundo subplot: gráfico de dispersão
ax2.scatter(x, y2, color='g')
ax2.set_title('Subplot 2: Função Cosseno')
ax2.set_xlabel('x')
ax2.set_ylabel('Cosseno(x)')

# Ajustando o layout e exibindo os subplots
plt.savefig('subplots.png', format='png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

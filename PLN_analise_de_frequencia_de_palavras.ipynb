# Abrindo e entendendo os datasets

## Dataset acadêmico

```python
dataset_academico = load_dataset("capes")
```

```python
dataset_academico['train']['translation']
```

```python
dataset_academico = [texto['pt'] for texto in dataset_academico['train']['translation']]
```

## Dataset e-commerce

```python
dataset_ecommerce = load_dataset("ruanchaves/b2w-reviews01")
```

```python
dataset_ecommerce
```

```python
dataset_ecommerce['train']['review_text']
```

```python
dataset_ecommerce = [texto for texto in dataset_ecommerce['train']['review_text'] if isinstance(texto, str)]
```

# Pré-processamento dos textos

```python
dataset_academico
```

```python
documentos = {}
```

```python
documentos['academico'] = " ".join(dataset_academico)
```

```python
documentos['academico'][:5000]
```

```python
documentos['ecommerce'] = " ".join(dataset_ecommerce)
```

```python
documentos['ecommerce'][:5000]
```

```python
def pre_process(text):
    text = text.replace(",", "")
    text = text.replace(".", "")
    text = text.replace("(", "")
    text = text.replace(")", "")
    text = text.replace("!", "")
    text = text.replace("?", "")
    text = text.lower()
    text = text.split(" ")
    return text
```

```python
documentos['academico'] = pre_process(documentos['academico'])
documentos['ecommerce'] = pre_process(documentos['ecommerce'])
```

```python
documentos['academico'][:20]
```

```python
documentos['ecommerce'][:20]
```

# Contando tokens

```python
contador_frequencia = {
    'academico': {},
    'ecommerce': {}
    }

# Um só loop tanto para o gênero acadêmico quanto para o gênero ecommerce
for genero in contador_frequencia:
    documento = documentos[genero]
    for token in documento:
        if not token in contador_frequencia[genero]:
            contador_frequencia[genero][token] = 0
        contador_frequencia[genero][token] += 1
```

```python
contador_frequencia['academico']
```

```python
contador_frequencia['ecommerce']
```

# Frequência proporcional

```python
documentos['academico']
```

```python
qtd_tokens = {
    'academico': len(documentos['academico']),
    'ecommerce': len(documentos['ecommerce'])
    }
```

```python
qtd_tokens
```

```python
contador_frequencia
```

```python
freq_proporcional = {
    'academico': {},
    'ecommerce': {}
    }

# Um só loop para os dois gêneros
for genero in contador_frequencia:
    for token in contador_frequencia[genero]:
        freq_proporcional[genero][token] = contador_frequencia[genero][token] / qtd_tokens[genero]
```

```python
freq_proporcional['academico']
```

```python
freq_proporcional['ecommerce']
```

# Interpretando resultados

```python
freq_proporcional
```

```python
ranking_academico = sorted(freq_proporcional['academico'].items(), key=lambda x: x[1], reverse=True)
ranking_ecommerce = sorted(freq_proporcional['ecommerce'].items(), key=lambda x: x[1], reverse=True)
```

```python
ranking_academico[:20]
```

```python
ranking_ecommerce[:20]
```

```python
exclusivo_academico = set(documentos['academico']) - set(documentos['ecommerce'])
```

```python
exclusivo_academico
```

```python
exclusivo_ecommerce = set(documentos['ecommerce']) - set(documentos['academico'])
```

```python
exclusivo_ecommerce
```

```python
ranking_academico_exclusivo = sorted({x for x in freq_proporcional['academico'].items() if x[0] in exclusivo_academico}, key=lambda x: x[1], reverse=True)
ranking_ecommerce_exclusivo = sorted({x for x in freq_proporcional['ecommerce'].items() if x[0] in exclusivo_ecommerce}, key=lambda x: x[1], reverse=True)
```
